```python
import numpy as np
import matplotlib.pyplot as plt
```

# Notebook2
This notebook creates a plot of the function
$$
f(x) = \frac{x^2}{2}
$$
on the domain $x \in [0,1]$.

```python
n_points = 100
x = np.linspace(0,1,n_points)
y = x**2/2
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("f(x)")
plt.show()
```
