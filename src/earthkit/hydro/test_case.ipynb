```python
import numpy as np
import earthkit.hydro as ekh

net = ekh.river_network.load("efas", "5", use_cache=False)
v = np.random.rand(net.n_nodes)*100


min_dist_to_sink = ekh.length.array.to_sink(net, path='shortest')
max_dist_to_source = ekh.length.array.to_source(net, path='longest')
upstream_sum = ekh.upstream.array.sum(net, v)
weighted_upstream_sum = ekh.upstream.array.sum(net, v, node_weights=min_dist_to_sink)
numerator = weighted_upstream_sum - min_dist_to_sink * upstream_sum
denominator = max_dist_to_source * upstream_sum
result = numerator/denominator

assert (~(result>1)).all()
assert (~(result<0)).all()
```
